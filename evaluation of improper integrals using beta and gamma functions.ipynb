verify that beta(m,n)= gamma(m)gamma(n)/gamma(m+n) for m=5 and n=7

```python
from sympy import beta , gamma
m= 5;
n= 7;
m=float (m);
n=float (n);
s= beta (m,n);
t= (gamma (m)* gamma (n))/ gamma (m+n);
print ('s = %0.4f' %s)
print ('t = %0.4f' %t)
if(abs (s-t)<=0.00001):
  print ('s=t (beta and gamma are related) ')
else :
  print('giveen values are wrong')
```

1. apply trapezoidal rule to evaluate ∫(0,1) 1/1+x^2 dx taking 6 equal parts

```python
def my_func(x):
  return 1/(1+x**2)
def trapezoidal (a, b, n):
  h= (b-a)/ n
  for i in range (1 , n):
   k = a + i * h
   integration =integration+  2 * my_func(k)
  integration = integration * h/2
  return integration
lower_limit = float(input('enter lower limit'))
upper_limit = float(input('enter upper limit'))
sub_intervals = int(input('enter number of sub intervals'))
result = trapezoidal(lower_limit, upper_limit, sub_intervals)
print("integral value by trapezoidal rule is:%0.4f"%result)
```

2. ∫(0,3.14) sin^2 x dx taking 6 equal parts

```python
from sympy import *
def my_func(x):
    return (sin(x)**2)
def trapezoidal (a, b, n):
    h = (b - a) / n
    integration = 0
    for i in range(1, n):
        k = a + i * h
        integration += 2 * my_func(k)
    integration = (integration + my_func(a) + my_func(b)) * h / 2
    return integration

lower_limit = float(input('Enter lower limit: '))
upper_limit = float(input('Enter upper limit: '))
sub_intervals = int(input('Enter number of sub intervals: '))
result = trapezoidal(lower_limit, upper_limit, sub_intervals)
print("Integral value by trapezoidal rule is: %0.4f" % result)
```

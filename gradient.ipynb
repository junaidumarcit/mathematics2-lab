3) prove that the vector (yz-x^2)i + (4y-(z^2)x)j + (2xz-4z)k is solenoidal

```python
from sympy.physics.vector import *
from sympy import var
var('x,y,z')
v= ReferenceFrame ('v')
F=((v[1]*v[2]-v[0]**2))*v.x + (4*v[1]-(v[2]**2)*v[0])*v.y + (2*v[0]*v[2]-4*v[2])*v.z
G=divergence (F,v)
F=F.subs([(v[0],x),(v[1],y),(v[2],z)])
print("given scalar function F=")
display(F)
G=G.subs([(v[0],x),(v[1],y),(v[2],z)])
print("Divergence of F=")
display(G)
```

4)Find the curl of F=[xy^2]i + [2(x^2)yz]j - [3yz^2]k

```python
from sympy.physics.vector import *
from sympy import var
var('x,y,z')
v= ReferenceFrame ('v')
F=(v[0]*v[1]**2)*v.x + 2*((v[0]**2)*v[1]*v[2])*v.y - (3*v[1]*v[2]**2)*v.z
G=curl (F,v)
F=F.subs([(v[0],x),(v[1],y),(v[2],z)])
print("given vector function F=")
display(F)
G=G.subs([(v[0],x),(v[1],y),(v[2],z)])
print("Curl of F=")
display(G)
```

6) if u=x+y+z, v=x^2+y^2+z^2, w=yz+zx+xy, find grad u, grad v, grad z

```python
from sympy.physics.vector import *
from sympy import var
var('x,y,z')
v= ReferenceFrame ('v')
u=v[0]+v[1]+v[2]
Gu=gradient(u,v)
u=u.subs([(v[0],x),(v[1],y),(v[2],z)])
print("given scalar function u=")
display(u)
Gu=Gu.subs([(v[0],x),(v[1],y),(v[2],z)])
print("gradient of u=")
display(Gu)

V=v[0]**2+v[1]**2+v[2]**2
GV=gradient(V,v)
V=V.subs([(v[0],x),(v[1],y),(v[2],z)])
print("given scalar function V=")
display(V)
GV=GV.subs([(v[0],x),(v[1],y),(v[2],z)])
print("gradient of V=")
display(GV)

w=v[1]*v[2]+v[2]*v[0]+v[0]*v[1]
Gw=gradient(w,v)
w=w.subs([(v[0],x),(v[1],y),(v[2],z)])
print("given scalar function w=")
display(w)
Gw=Gw.subs([(v[0],x),(v[1],y),(v[2],z)])
print("gradient of w=")
display(Gw)
```

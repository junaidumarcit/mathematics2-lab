find the center of gravity of cardiod a(1+cos(theta))

```python
from sympy import *
r,theta,a=symbols('r theta a')
I1=integrate(cos(theta)*r**2,(r,0,a*(1+cos(theta))),(theta,-pi,pi))
I2=integrate(r,(r,0,a*(1+cos(theta))),(theta,-pi,pi))
I3=I1/I2
pprint(I3)
I4=integrate(sin(theta)*r**2,(r,0,a*(1+cos(theta))),(theta,-pi,pi))
I5=integrate(r,(r,0,a*(1+cos(theta))),(theta,-pi,pi))
I6=I4/I5
pprint(I6)
```

8.find the center of gravity of one loop of r=asin2theta

```python
from sympy import *
r,theta,a=symbols('r theta a')
I1=integrate(cos(theta)*r**2,(r,0,a*(sin(2*theta))),(theta,0,pi/2))
I2=integrate(r,(r,0,a*(sin(2*theta))),(theta,0,pi/2))
I3=I1/I2
pprint(I3)
I4=integrate(sin(theta)*r**2,(r,0,a*(sin(2*theta))),(theta,0,pi/2))
I5=integrate(r,(r,0,a*(sin(2*theta))),(theta,0,pi/2))
I6=I4/I5
pprint(I6)
```

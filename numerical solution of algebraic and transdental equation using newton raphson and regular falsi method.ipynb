Newton Raphson method

1)FIND THE ROOT OF THE EQUATION 3X=COSX+1,Near x=1 by N-R M

```python
from sympy import *
x=symbols('x')
g=input('enter the function')
f=lambdify(x,g)
dg=diff(g);
df=lambdify(x,dg)
x0=float(input('enter initial approximation:'))
N=int(input('enter the number of iterations:'))
for i in range(1,N+1):
  x1=(x0-(f(x0)/df(x0)))
  print('iteration %d \t the root %0.3f \t function value %0.3f \n' %(i,x1,f(x1)));
  x0=x1
print('final value of the root is %0.3f' %x1)
```
